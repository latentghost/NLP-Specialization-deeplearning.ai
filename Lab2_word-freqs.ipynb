```python
import nltk
from nltk.corpus import twitter_samples
import matplotlib.pyplot as plt
import numpy as np
```

```python
import ssl
try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context
```

```python
nltk.download('twitter_samples')
```

```python
all_pos_tw = twitter_samples.strings('positive_tweets.json')
all_neg_tw = twitter_samples.strings('negative_tweets.json')
```

```python
from utils import process_tweet,build_freqs
nltk.download('stopwords')
```

```python
tweets = all_pos_tw + all_neg_tw
```

```python
labels = np.append(np.ones((len(all_pos_tw))),np.zeros((len(all_neg_tw))))
```

```python
tweets_processed = []
for tweet in tweets:
    tweets_processed.append(process_tweet(tweet=tweet))
```

```python
freqs = build_freqs(tweets=tweets_processed,ys=labels)
```

```python
X = []

for tweet in tweets_processed:
    pos = 0
    neg = 0

    for word in tweet:
        pos += freqs[word][0]
        neg += freqs[word][1]
    
    X.append([1,pos,neg])
```
